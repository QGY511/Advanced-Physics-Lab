```python
file=open("Muon_Data_5_11.txt","r" )
event=[]
file=file.readlines()
for line in file:
    data=line.split()
    event.append(float(data[1]))
    
bins=np.arange(0,4096,4096/30)
plt.xlabel('ADC chanel',fontsize=18)
plt.ylabel('Number of events',fontsize=18)
plt.hist(event, bins, histtype='bar', rwidth=0.5,label="data")

counts, edges = np.histogram(event,bins=30,range= (0,4096))
errs=counts**0.5
cbins=(edges[:-1]+edges[1:])/2.0


def exp_func(x,a,lamb,b):
    return a*np.exp(-lamb*x)+b



guess_a=1.0
guess_lamb=0.0
guess_b=0.0


par, cov = optimize.curve_fit(exp_func, cbins, counts, p0=[guess_a, guess_lamb, guess_b],sigma=errs,absolute_sigma=True)
unc=np.sqrt(np.diag(cov))

unc_a=unc[0]
unc_lamb=unc[1]
unc_b=unc[2]


fit_a=par[0]
fit_lamb=par[1]
fit_b=par[2]

xf=np.linspace(0,4096,2000)
yf=fit_a*np.exp(-fit_lamb*xf)+fit_b

plt.plot(xf,yf,"r--",label="line fit")
chi_square=0
for i in range(len(cbins)):
    chi_square=chi_square+(fit_a*np.exp(-fit_lamb*cbins[i])+fit_b-counts[i])**2/errs[i]**2

    
plt.errorbar(cbins,counts , yerr=errs,fmt="ko",label="error bar")

plt.legend()


print(1/fit_lamb*10/4096)
print(1/(fit_lamb+unc_lamb)*10/4096-1/fit_lamb*10/4096)

print(fit_b)
print(unc_b)

print(fit_a)
print(unc_a)

print(chi_square/27)
```
